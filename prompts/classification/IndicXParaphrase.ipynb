```python
from groq import Groq
client = Groq(
    api_key="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",
)
def get_labels(sentence1,sentence2):
  prompt = f"Sentence 1: {sentence1} Sentence 2: {sentence2} Question: Does Sentence 1 paraphrase Sentence 2 ? Yes(1) or No(0)?. Provide the answer only as a number."
  chat_completion = client.chat.completions.create(
      messages=[
          {
              "role": "user",
              "content": prompt,
          }
      ],
      model="llama3-8b-8192",
  )
  return chat_completion.choices[0].message.content
```

```python
predictions = []

# Make predictions for each text in the processed DataFrame
for index,row in eval_df.iterrows():
    label = get_labels(row["sentence1"],row["sentence2"])
    predictions.append(label)

# Create a new DataFrame with the original text and the predicted labels
predicted_df = pd.DataFrame({
    'sentence1': eval_df['sentence1'],
    'sentence2': eval_df['sentence2'],
    'label': eval_df['label'],
    'predicted_label': predictions
})
predicted_df
```

```python
predicted_df['correct'] = predicted_df['label'].astype(int) == predicted_df['predicted_label'].astype(int)
accuracy = predicted_df['correct'].mean()

print(f'Accuracy: {accuracy * 100:.2f}%')
```
